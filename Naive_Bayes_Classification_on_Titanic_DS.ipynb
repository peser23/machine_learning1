```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/drive/MyDrive/data/titanic.csv")
df
```

```python
#drop unused columns
df.drop(columns=['PassengerId','Name','Ticket','Cabin','SibSp','Parch','Embarked'], axis='columns', inplace=True)
df
```

```python
target = df.Survived
inputs = df.drop(columns=['Survived'], axis='columns')
```

```python
#Convert Sex column to male, female columns
dummies = pd.get_dummies(df.Sex)
#dummies
```

```python
#add the 2 new columnc to inputs
inputs = pd.concat([inputs, dummies], axis='columns')
inputs.drop(columns=['Sex'], axis='columns', inplace=True)
inputs
```

```python
#check for NULL values
inputs.isnull().sum()
```

```python
inputs["Age"] = inputs["Age"].fillna(inputs["Age"].mean())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size=0.20, random_state=7854)
```

```python
len(X_test)
```

### Start Naive Bayes Classifier

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
#get prediction for 1st 10 records of x_test
model.predict(X_test[:10])
```

```python
#get Predition Probability of Not Survived vs Survived for 1st 10 records of x_test
model.predict_proba(X_test[:10])
```
