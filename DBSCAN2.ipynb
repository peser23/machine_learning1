# DBSCAN (Example: 2) -

```python
from sklearn.cluster import DBSCAN
import numpy as np
from matplotlib import pyplot as plt

X = np.array([[1,2],[3,4],[4,3],[8,7],[7,8],[25,80]])
```

```python
plt.scatter(x=X[:,0],y=X[:,1])
```

## Run DBSCAN on above array

```python
cluster1 = DBSCAN(eps=3, min_samples=2)
cluster1.fit(X)

cluster1.labels_
```
