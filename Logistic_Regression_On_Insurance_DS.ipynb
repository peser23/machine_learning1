```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/data/insurance_data.csv')
df.head()
```

```python
#plot the data using scatter  pot
plt.scatter(df['age'], df['bought_insurance'], color='red', marker='+')
plt.xlabel('Age')
plt.ylabel('Bought Insurance?')
```

```python
df.shape
```

```python
# Split the data to test, train
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[['age']], df['bought_insurance'], test_size=0.10)
```

#### Train the Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
