```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Read the csv to dataframe
df = pd.read_csv('/content/drive/MyDrive/data/homeprices.csv')
```

```python
#Plot the data
plt.xlabel('Area (in sq ft)')
plt.ylabel('price in US$')
plt.scatter(df['area'],df['price'],color='red',marker='+')
```

```python
#train the Linear Regression Model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
X = df[['area']]
model.fit(X,df['price'])
```

```python
#test the model
model.predict(X=[[3300]])
```

```python
#Get the Cofiecient and intercept
m = model.coef_
b = model.intercept_
print(m)
print(b)
```

```python
#Plot the best fitted line
plt.xlabel('Area (in sq ft)')
plt.ylabel('price in US$')
plt.scatter(df['area'],df['price'],color='red',marker='+')
plt.plot(df['area'], model.predict(df[['area']]))
```
